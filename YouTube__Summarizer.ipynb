```python
# Youtube Summarizer
# Rishabh Nirmalkar


import os


from google import genai
from google.genai import types

def main():
  api_key = os.getenv('GEMINI_API_KEY', 'AIzaSyCycjW6jmMfeJP7YLZq5kWDhmFcUk-D4LE')
  client = genai.Client(api_key=api_key)
  youtube_url = input("Enter Youtube video URL:").strip()
  if not youtube_url:
    print("No Youtube url provided. Exiting.")
    return
  prompt_text=(
        "You will analyze and summerize the content of the following YouTube video."
        "provide a clear, consise summary focusing on main topic, key takeways, "
        "and important points."
  )
  content = types.Content(
        parts=[
            types.Part(text=prompt_text),
            types.Part(file_data=types.FileData(file_uri=youtube_url))
            ]
  )
  print(" Processing video...   this may take a moment")
  response = client.models.generate_content(
        model='gemini-2.5-pro',
        contents=content
  )
  print("\n===== video summary =====\n")
  print(response.text)
  print("\n=========================\n")
if __name__ == "__main__":
        main()
```
